# Using USEEIO-State
[USEEIO-State](https://github.com/USEPA/USEEIO-State) is a public github repository for accessing and analyzing State EEIO results in R. It includes scripts for building models, downloading already created models, and generating state Consumption Based Emission (CBE) results.

If using RStudio, these scripts can be knit with parameters directly within RStudio after cloning the repository:

image.png


*Note: Select Runtime -> Change runtime type --> Change "Python 3" to "R"*

```r
# Clones the repository into your local colab environment
system("git clone https://github.com/USEPA/USEEIO-State.git")
```

```r
# Set the working directory to the repository
setwd("USEEIO-State")
```

```r
# Install required supporting packages for generating model outputs
install.packages("rmarkdown")
install.packages("bookdown")
install.packages("officedown")
```

See the [examples folder](https://github.com/USEPA/USEEIO-State/tree/main/examples) for a list of available scripts.
1. **DownloadandSaveModelLocally**. Select and download an EPA-built USEEIO State model. Save the model locally as .rds for quick retrieval for further uses. Optionally write the model out to Excel.  

2. **BuildandSaveModelsLocally**. Select and build USEEIO State models from specs and save the model locally as .rds for quick retrieval for further uses. Optionally write the models out to Excel.

4. **StateCBE**. Calculate and display consumption based GHG emissions for a selected state, rendered as a Microsoft Word file.

## Download an Existing Model
Let's first download a pre-saved model. All published USEEIO-State models are stored as .rds files on [EPA's Data Commons](https://dmap-data-commons-ord.s3.amazonaws.com/index.html?prefix=#USEEIO-State/). See the [README](https://github.com/USEPA/USEEIO-State/blob/main/model_specs/README.md) for an explanation of the available versions.

```r
# Select the state(s), year(s), and model version of choice.
# Note not all years are available for all versions
my_params = list(states=c("VT"), years=c(2018:2022), model="v1.3-pecan")
rmarkdown::render("examples/DownloadandSaveModelLocally.Rmd", run_pandoc = FALSE, params=my_params)
```

## Build a Model
Using useeior, build a model directly from one of the available model specification files.

```r
my_params = list(states=c("NY"), years=c(2018:2019), model="v1.3-pecan")
rmarkdown::render("examples/BuildandSaveModelsLocally.Rmd", run_pandoc=FALSE, params=my_params)
```

## Construct a Consumption Based Emission Inventory
For any selected state, you can generate a dataset and brief report that highlights the consumption based emissions for the time series.

_Note: running this in colab can take 10-15 minutes_

```r
my_params = list(state="VT", years=c(2018:2022), model="v1.3-pecan")
rmarkdown::render("examples/StateCBE.Rmd", run_pandoc = FALSE, params=my_params)
```
